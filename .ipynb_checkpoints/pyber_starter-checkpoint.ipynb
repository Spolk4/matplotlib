```python
Three observable trends in the data:
    #1. Urban areas tend to use the service more either do to population or the lack of POVs
    #2. Urban areas also have the benefit of reduced average fare per ride either due to shorter 
        #distance traveled or the incredible amount of drivers available competing for fare
    #3. There is an observable correlation between the decreasing amount of rides per city type 
        #and the increase in average fare for all city types.
```

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

#view data
#city_data.head()
#ride_data.head()

# Combine the data into a single dataset
merged_data_df = pd.merge(city_data, ride_data, on = "city", how = "left")

# Display the data table for preview
merged_data_df.head()
```

```python
Resi_types= merged_data_df["type"].unique()
Resi_types
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_df =merged_data_df.loc[merged_data_df['type']=='Urban'].groupby([merged_data_df["city"]])
suburban_df =merged_data_df.loc[merged_data_df['type']=='Suburban'].groupby([merged_data_df["city"]])
rural_df =merged_data_df.loc[merged_data_df['type']=='Rural'].groupby([merged_data_df["city"]])
#urban_df.head()
urban_x = urban_df["ride_id"].count()
urban_y=urban_df["fare"].mean()
urban_size = urban_df["driver_count"].mean()

suburban_x = suburban_df["ride_id"].count()
suburban_y = suburban_df["fare"].mean()
suburban_size = suburban_df["driver_count"].mean()

rural_x = rural_df["ride_id"].count()
rural_y = rural_df["fare"].mean()
rural_size = rural_df["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, label = "Urban",s=urban_size*10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(suburban_x, suburban_y, label = "Suburban", s=suburban_size*10, color = ["gold"], edgecolor = "black", alpha = 0.75, marker = "o")
plt.scatter(rural_x, rural_y, label = "Rural", s=rural_size*10, color = ["lightskyblue"], edgecolor = "black", alpha = 0.75, marker= 'o')

# Incorporate the other graph properties
plt.grid()
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")
# Create a legend
legend = plt.legend(fontsize=8, title = "Type of City", loc = "best")
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,30,"Note: Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("PyberRideSharingData2016.png", bbox_inches='tight')
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents 
cityTypes_grouped = merged_data_df.groupby(['type'])
fare_sum = cityTypes_grouped['fare'].sum()

# Build Pie Chart
labels_type = ['Rural', 'Suburban','Urban']
explode =(0,0,0.1)
colors = ["lightskyblue", "gold", "lightcoral"]
plt.title("% of Total Fares by City Type")
plt.pie(fare_sum, explode = explode, labels = labels_type, colors = colors, autopct ="%2.2f%%", shadow = True, startangle = 140)

# Save Figure
plt.savefig("TotalFaresbyCityType.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_count = cityTypes_grouped["ride_id"].count()

# Build Pie Chart
labels_type = ['Rural', 'Suburban','Urban']
explode =(0,0,0.1)
colors = ["lightskyblue", "gold", "lightcoral"]
plt.title("% of Total Rides by City Type")
plt.pie(rides_count, explode = explode, labels = labels_type, colors = colors, autopct ="%2.2f%%", shadow = True, startangle = 145)

# Save Figure
plt.savefig("%ofTotalRidesbyCityType.png")

# Show Figure
plt.show()
```

```python
# Calculate Driver Percents: use city_data.csv
type_drivers_grp = city_data.groupby(['type'])
drivers_grp_sum = type_drivers_grp['driver_count'].sum()

# Build Pie Charts
labels_type = ['Rural', 'Suburban','Urban']
explode =(0,0,0.1)
colors = ["lightskyblue", "gold", "lightcoral"]
plt.title("% of Total Drivers by City Type")
plt.pie(drivers_grp_sum, explode = explode, labels = labels_type, colors = colors, autopct ="%2.2f%%", shadow = True, startangle = 145)

# Save Figure
plt.savefig("%ofTotalDriversbyCityType.png")

# Show Figure
plt.show()
```
